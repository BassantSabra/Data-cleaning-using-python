```python
# Data connection
folder_path = "/content/drive/MyDrive/Data for carerha internship/Cleaning Data - Practice"
files = os.listdir(folder_path)
files
# Data cleaning and transformation
dfs=[]
for file in files:
  file_path=folder_path +"/"+ file
  print(file)
  df = pd.read_excel(file_path)
  # remove null values and unwanted cols
  df = df.dropna(how = "all")
  df = df.drop("Unnamed: 5",axis = 1)
  # make conditional column using numpy
  df["date"] = np.where( df["Daily Production Report"] =="Date", df["Unnamed: 1"],np.nan)
  #  fill date col
  df["date"] = df["date"].ffill()
  # remove (date row)
  df = df[df["Daily Production Report"] !="Date"].reset_index(drop = True)
  # # fill na with project_name
  df["Daily Production Report"] =  df["Daily Production Report"].fillna("project_name")
  #  make headwe as first row  or ( it can be done in one step without reset_index using (df.columns=df.iloc[0])
  df.columns = df.loc[0]
  # remove the first row
  df = df.drop(0)
  #  rename the last col
  df = df.rename ({df.columns[-1]:"date"} ,axis = 1)
  #  making unpivotable
  df = pd.melt(df,id_vars= ["project_name","date"], value_name= "quantity" , var_name = "product")
  dfs.append(df)
  # print(df)


#  concat files
df =pd.concat(dfs)
# cleaning data
df["quantity"]= df["quantity"].astype(int)
# finial dataset
df.info()
df
```
